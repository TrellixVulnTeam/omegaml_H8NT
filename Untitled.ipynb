```python
def paginate(coll, cond={}, size=100, key='_id'):
    """
    return one set of objects at a time
    
    Args:
        coll (Collection): the Collection object to pageinate
        cond (dict): the filter conditions
        size (int): the size of the page in #rows
        key (string): the column to use as the key
        
    Issues:
        what if I want to sort results?
    """
    cursor = coll.find(cond)
    row_count = cursor.count()
    last_id = None
    while row_count:
        if last_id is not None:
            page_cond = {key: {
                '$gt': last_id
            }}
        else:
            page_cond = {}
        page_cond.update(cond)
        page_cursor = coll.find(page_cond)
        result = [row for row in page_cursor.limit(size)]
        last_id = result[-1][key]
        yield result
        row_count -= len(result)

total = 0
for i, page in enumerate(paginate(coll, size=10000)):
    print len(page), page[-1]
    total += len(page)
print total
```

```python
import re
name = 'foo[]'

all_columns = ['d', 'a', 'b']

def columnset(specs, all_columns):
    """
    find the specified columns in data[colspec] in list of all columns
    
    colspec can be any of
    
    * a single column, e.g. foo[a]
    * a comma separated list of columns, e.g. foo[a,b]
    * an open-ended slice, e.g. foo[a:] => all columns following a, inclusive
    * an closed slice, e.g. foo[a:b] => all columns between a,b, inclusive
    * a close-ended slice, e.g. foo[:b] => all columns up to b, inclusive
    * an empty slice, e.g. foo[:] => all columns
    """
    colspec_pattern = re.compile(r"(?P<name>.*)\[(?P<colspec>.*)\]$")
    match = colspec_pattern.match(specs)
    columns = []
    if match is not None:
        specs, colspec = match.groups()
        if ':' in colspec:
            from_col, to_col = colspec.split(':')
            from_i = all_columns.index(from_col) if from_col else 0
            to_i = (all_columns.index(to_col) if to_col else len(all_columns)) + 1
            columns = all_columns[from_i:to_i]
        elif '^' in colspec:
            columns = [col for col in all_columns if col not in colspec.split(',')]
        else:
            columns = colspec.split(',')
    return columns

columnset('foo[^b]', all_columns)
```
